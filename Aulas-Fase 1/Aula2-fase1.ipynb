```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
#preparando dados
x = torch.tensor([
    [5.0], [20.0], [10.0], [5.0], [10.0],
    [5.0], [10.0], [10.0], [5.0], [10.0],
    [5.0], [10.0], [10.0], [5.0], [10.0],
    [5.0], [10.0], [10.0], [5.0], [10.0]
    ], dtype=torch.float32)

#resultado esperado
y = torch.tensor([
    [30.5], [63.0], [67.0], [29.0], [62.0],
    [30.5], [63.0], [67.0], [29.0], [62.0],
    [30.5], [63.0], [67.0], [29.0], [62.0],
    [30.5], [63.0], [67.0], [29.0], [62.0]
    ], dtype=torch.float32)
```

```python
#construindo rede neural
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(1, 5)  # Camada oculta com 10 neurônios
        self.fc2 = nn.Linear(5, 1)   # Camada de saída com 1 neurônio

    def forward(self, x):
        x = torch.relu(self.fc1(x))  # Função de ativação ReLU
        x = self.fc2(x)              # Saída linear
        return x

model = Net()

criterion = nn.MSELoss()  # Função de perda (Mean Squared Error)
optimizer = optim.SGD(model.parameters(), lr=0.01)  # Otimizador
```

```python
#treinamento
for epoch in range(1000):
    optimizer.zero_grad()  # Zera os gradientes
    outputs = model(x)     # Forward pass
    loss = criterion(outputs, y)  # Calcula a perda
    loss.backward()        # Backward pass
    optimizer.step()       # Atualiza os pesos


    if epoch % 100 == 99:  # Imprime a perda a cada 100 épocas
        print(f'Epoch [{epoch+1}/1000], Loss: {loss.item():.4f}')
```

```python
with torch.no_grad():
    predicted = model(torch.tensor([[10.0]], dtype=torch.float32))
    print(f'Previsao de tempo: {predicted.item():.2f} minutos')
```
