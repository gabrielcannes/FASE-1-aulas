```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
composto1 = [1,1,1]
composto2 = [0,0,0]
composto3 = [1,0,1]
composto4 = [0,1,0]
composto5 = [1,1,0]
composto6 = [0,0,1]

dados_treino = [composto1, composto2, composto3, composto4, composto5, composto6]
rotulos_treino = ['S', 'N', 'S', 'N', 'S', 'S']


modelo = LinearSVC()
# Treina o modelo com os dados de treino
modelo.fit(dados_treino, rotulos_treino)

# Testes de validacao
teste1 = [1,0,0]
teste2 = [0,1,1]
teste3 = [1,0,1]

dados_teste = [teste1, teste2, teste3]
```

```python
previsoes = modelo.predict(dados_teste)

mapeamento_previsoes = {'S': 'Soluvel', 'N': 'Insolúvel'}

print("Previsoes do modelo para os dados testados:", previsoes)
for i, previsao in enumerate(previsoes):
    print(f'O composto {i+1} é {mapeamento_previsoes[previsao]}')
```

```python
#Gabarito para que eu possa descobrir qual a taxa de acerto do modelo
rotulos_teste = ['S', 'S', 'S']

previsoes = modelo.predict(dados_teste)

taxa_acerto = accuracy_score(rotulos_teste, previsoes)
print(f'Taxa de acerto do modelo: {taxa_acerto*100:.2f}%')
```
